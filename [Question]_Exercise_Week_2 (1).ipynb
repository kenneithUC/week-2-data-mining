7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Encode the 'role' column
label_encoder = LabelEncoder()  # Create an instance of LabelEncoder
df['role'] = label_encoder.fit_transform(df['role'])
# Separate features and target variable
x=df.drop(['hero_name','win_rate','pick_rate','ban_rate','release_year','role'],axis=1)
y=df['role']
# Split the data into training and testing sets
df.split=train_test_split(x,y,test_size=0.2,random_state=10)
# Create a Decision Tree Classifier
df.clf=DecisionTreeClassifier()
# Train the model
x_train,x_test,y_train,y_test=df.split
df.clf.fit(x_train,y_train)
# Make predictions on the test set
df.y_pred=df.clf.predict(x_test)
# Evaluate the model
df.accuracy=accuracy_score(y_test,df.y_pred)
print("Accuracy:",df.accuracy)
# Predict the role for the given character
prediction=df.clf.predict([[6.0,7.0,4.0,3.0,250,9,380,37,119,24,2689,0.75,16.0]])
# Get the predicted role (dicoding)
predicted_role = label_encoder.inverse_transform(prediction)
print("Predicted role:", predicted_role[0])
```
